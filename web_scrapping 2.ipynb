```python
from bs4 import BeautifulSoup
import requests
url = BeautifulSoup('http://www.nepalstock.com/floorsheet', 'html.parser')
response = requests.get(url)
data = response.text
data = BeautifulSoup(data,'lxml')
data = data.table
tags = data.find_all('tr')
tags = tags[1:-3]
l = []
for i in tags:  
    l.append(i.text)
output = []
for i in range(len(l)):
    x = l[i].split('\n')
    y = x[1:-1]
    output.append(y)
print(output)
```

```python
import csv   
fields=output
with open('stock.csv', 'a') as f:
    x = csv.writer(f)
    for i in fields:
        x.writerow(i)
```

```python
from bs4 import BeautifulSoup
import requests
n = int(input("Enter n = "))
output = []
c = 1
for z in range(1,n+1):
 
    url = BeautifulSoup(f'http://www.nepalstock.com/main/floorsheet/index/{z}/', 'html.parser')
    response = requests.get(url)
    data = response.text
    data = BeautifulSoup(data,'lxml')
    data = data.table
    tags = data.find_all('tr')
    tags = tags[1:-3]
    l = []
    for i in tags:  
        l.append(i.text)
  
    for i in range(len(l)):
        x = l[i].split('\n')
        y = x[1:-1]
        output.append(y)

print(output)
```

```python
x = ['S.N.',
 'Contract No',
 'Stock Symbol',
 'Buyer Broker',
 'Seller Broker',
 'Quantity',
 'Rate',
 'Amount']
y = []
for i in output:
    if i == x:
        continue
    y.append(i)
print(y)
```

```python
final_output = [x]+y
final_output
```

```python
import csv   
fields=final_output
with open('stock.csv', 'w') as f:
    x = csv.writer(f)
    for i in fields:
        x.writerow(i)
```
